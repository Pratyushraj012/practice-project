```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
X=df.drop('Item_Outlet_Sales',axis=1)
y=df.Item_Outlet_Sales
```

```python
scaler=StandardScaler()
```

```python
x_scaled=scaler.fit_transform(X)
```

```python
x_scaled.shape[1]
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
vif=pd.DataFrame()
vif["vif"]=[variance_inflation_factor(x_scaled,i)
           for i in range(x_scaled.shape[1])]
vif["feature"]=X.columns
vif
```

```python
## Removing outlet_Location_type
df.drop('Outlet_Location_Type',axis=1,inplace=True)
df2.drop('Outlet_Location_Type',axis=1,inplace=True)
```

```python
X=df.drop('Item_Outlet_Sales',axis=1)
y=df.Item_Outlet_Sales
```

```python
x_scaled=scaler.fit_transform(X)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.30,random_state=41)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
ypred=lr.predict(x_train)
```

```python
from  sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
```

```python
r2_score(y_train,ypred)
```

```python
y_pred=lr.predict(x_test)
r2_score(y_test,y_pred)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cross_val=cross_val_score(lr, x_test,y_test,cv=5)
cross_val.mean()
```

```python
from xgboost import XGBRegressor
xgb=XGBRegressor()
xgb.fit(x_train,y_train)
```

```python
ypred=xgb.predict(x_train)
```

```python
r2_score(y_train,ypred)
```

```python
y_pred=xgb.predict(x_test)
```

```python
r2_score(y_test,y_pred)
```

```python
cross_val=cross_val_score(xgb, x_test,y_test,cv=5)
cross_val.mean()
```

Obsevation:- Xgboost is slightly better than linear regression but overfitting is high.
